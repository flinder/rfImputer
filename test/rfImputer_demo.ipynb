```python
import pandas as pd
import sys
sys.path.insert(1, '../')
from rfImputer import rfImputer

cols = ['Proof_cut', 'Viscosity', 'Caliper', 'Ink_temperature', 'Humifity',
        'Roughness', 'Blade_pressure', 'Varnish_pct', 'Press_speed', 'Ink_pct',
        'Solvent_pct', 'Esa_voltage', 'ESA_amperage', 'Wax', 'Hardener',
        'Roller_durometer', 'Density', 'Anode_ratio', 'Chrome_content', 'bands']
df = pd.read_csv('bands.csv')
df.columns = cols

df.replace('?', 'NaN', inplace = True)
df.replace('band', '1', inplace = True)
df.replace('noband', '0', inplace = True)

for col in df:
    if col == 'bands':
        continue
    df[col] = df[col].astype(float)
```

```python
df.head(20)
df.shape[0]
```

```python
# Instantiate Imputer with data
imp_df = rfImputer(df)
```

```python
imp_df.col_types
```

```python
# Print basic attributes
pprint(imp_df.prop_missing)
print imp_df.imputed_values
print imp_df.imputation_scores
```

```python
# Mean mode impute
imp_df.impute('simple')
imp_df.imputed_df().head(20)
```

```python
# Random Forest Impute
imp_df.impute('random_forest', {'n_estimators': 50, 'n_jobs': 4})
imp_df.imputed_df().head(20)
```
